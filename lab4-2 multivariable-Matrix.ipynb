# mutil-variable with Matrix
$$H(x)=\begin{bmatrix} { w }_{ 1 } & { w }_{ 2 } \end{bmatrix}\times \begin{bmatrix} { x }_{ 1 } \\ { x }_{ 2 } \end{bmatrix}+b={ w }_{ 1 }{ x }_{ 1 }+{ w }_{ 2 }{ x }_{ 2 }+b$$
---

### 1. 학습 데이터

```python
import tensorflow as tf

x_data = [[1.,0.,3.,0.,5.],
           [0.,2.,0.,4.,0.]]
y_data = [1,2,3,4,5]
```

### 2. variable

```python
W = tf.Variable(tf.random_uniform([1,2],-1.0,1.0))
b = tf.Variable(tf.random_uniform([1],-1.0,1.0))
```

### 3. Hypothesis
$$H({ x }_{ 1 },{ x }_{ 2 })={ w }_{ 1 }{ x }_{ 1 }+{ w }_{ 2 }{ x }_{ 2 }+b$$


```python
hypothesis = tf.matmul(W,x_data) + b
```

### 4. cost

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

### 5. minimise cost

```python
a = tf.Variable(0.1)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)
```

### 6. 초기화

```python
init = tf.initialize_all_variables()
```

### 7. 세션 생성

```python
sess = tf.Session()
sess.run(init)
```

### 8. 학습 실행

```python
for step in range(10001):
    sess.run(train)
    if step %100 == 0:
        print(step,"- cost:",sess.run(cost),"| W:",sess.run(W),"| b:",sess.run(b))
```
